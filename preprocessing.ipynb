```python
import time
from datetime import datetime

df['time']='def'
for i in range(1,14313):
  x = df.loc[i]
  date = x['Date']
  Time = x['Time']
  if type(Time)==str and type(date)==str:
    datetime_str = date + ' ' + Time
    datetime_obj = datetime.strptime(datetime_str, '%d-%m-%y %H:%M')
    unixtime = time.mktime(datetime_obj.timetuple())
    x['time'] = unixtime
    df.loc[i] = x
  elif type(Time)!=str and type(date)==str:
    datetime_str = date
    datetime_obj = datetime.strptime(datetime_str, '%d-%m-%y')
    unixtime = time.mktime(datetime_obj.timetuple())
    x['time'] = unixtime
    df.loc[i] = x
  
```
